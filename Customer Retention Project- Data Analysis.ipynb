```python
# importing the libraries for visualization 
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# ploting for all columns and thier data distribution 
for col in df:
    plt.figure(figsize=(20,5))
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.title(col)
    sns.countplot(df[col])
    print("unique buckets:",len(df[col].unique()))
    print("\n")
    print("unique buckets %:",(df[col].value_counts()/len(df[col]))*100)
    plt.show()
   
```
