# 1)Write a python program to display all the header tags from wikipedia.org


```python
# send a request to server
wiki_requests=requests.get('https://www.wikipedia.org/')
wiki_requests
#get the contetst
wiki_contents=BeautifulSoup(wiki_requests.content)
wiki_contents
#get the header from wiki.org

headings=[]

for i in wiki_contents.find_all('div',class_='other-project-text'):
    headings.append(i.text.split()[0])
    
headings

```

# 2)Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release)and make data frame.


```python
IMDB_requests=requests.get('https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc')
IMDB_requests
IMDB_contents=BeautifulSoup(IMDB_requests.content)
IMDB_contents
```

```python
# get name of the movies
movie_name=[]

for i in IMDB_contents.find_all('h3',class_='lister-item-header'):
    movie_name.append(i.text.split('\n')[2])

    
movie_name
```

```python
Rating=[]

for i in IMDB_contents.find_all('div',class_='ratings-bar'):
    Rating.append(i.text.split('\n')[3])
    
Rating
```

```python
Release_year=[]

for i in IMDB_contents.find_all('span',class_='lister-item-year text-muted unbold'):
    Release_year.append(i.text.replace('(',' ').replace(')',' '))

    
Release_year
```

```python
print(len(Release_year),len(Rating),len(movie_name))
```

```python
import pandas as pd
IMDB_Top_100movies=pd.DataFrame({'Movie_Name':movie_name,'Rating':Rating,'Release_Year':Release_year})
IMDB_Top_100movies
```

# 3) Write a python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. name, rating, year ofrelease) and make data frame

```python
Indianmovies_Requests=requests.get('https://www.imdb.com/india/top-rated-indian-movies/')
Indianmovies_Requests
Indianmovies_contents=BeautifulSoup(Indianmovies_Requests.content)
Indianmovies_contents
```

```python
IND_Movies_Titles=[]

for i in Indianmovies_contents.find_all('td',class_='titleColumn'):
    IND_Movies_Titles.append(i.text.split('\n')[2])
    
IND_Movies_Titles
```

```python
IND_Movies_Rating=[]

for i in Indianmovies_contents.find_all('td','ratingColumn imdbRating'):
    IND_Movies_Rating.append(i.text.split('\n')[1])
    
IND_Movies_Rating
```

```python
IND_Movies_Releasedate=[]

for i in Indianmovies_contents.find_all('span','secondaryInfo'):
    IND_Movies_Releasedate.append(i.text.replace('(',' ').replace(')',' '))
    
IND_Movies_Releasedate
```

```python
IND_Top_100Movies=pd.DataFrame({'Movie_Title':IND_Movies_Titles,'Movie_Rating':IND_Movies_Rating,'Release_year':IND_Movies_Releasedate})
IND_Top_100Movies
```
